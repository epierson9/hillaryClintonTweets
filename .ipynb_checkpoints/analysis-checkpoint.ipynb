```python
"""
This is the code underlying the analysis of #WordsThatDontDescribeHillary. 
You will not be able to run the code since it relies on data + libraries that are private, but please contact me 
at emmap1@cs.stanford.edu if you would like the data or have comments, questions, or corrections.s
"""
import pandas as pd
import string
from collections import Counter
import numpy as np
import random
import twitterUtil as tu
import statsmodels.api as sm
pd.set_option('max_colwidth', 200)

def getWords(tweet):
    """
    This extracts the adjectives from tweet text, filtering out some of the most common confounds. 
    It doesn't work perfectly, since tweeters weren't kind enough to follow a perfectly consistent syntax
    but after some experimentation, I think this provides the best compromise of false positives / negatives. 
    """
    tweet = tweet.lower()
    tokens_to_remove = [a for a in tweet.split() if '@' in a or 'http' in a or 'fuck' in a or 'bern' in a or 'trump' in a] + ['feelthebern', 'tcot', 'p2', 'whatilearnedtoday', 'wordsthatdontdescribehillary', "wordsthatdon'tdescribehillary"]
    for token in tokens_to_remove:
        tweet = tweet.replace(token, '')
    tokens_to_split_on = ['\n', ',', '.', '&amp;', '#']
    for token in tokens_to_split_on:
        tweet = tweet.replace(token, '*****')
    tweet = tweet.split('*****')
    for i in range(len(tweet)):
        tweet[i] = tweet[i].translate(string.maketrans("",''), string.punctuation).strip()
    tweet = [a for a in tweet if len(a) > 0]
    return tweet
def bernie(x):
    try:
        x = x.lower()
        if 'bern' in x or 'sanders' in x:
            return True
        return False
    except:
        return False
def trump(x):
    try:
        x = x.lower()
        if 'dumptrump' in x or 'againsttrump' in x or 'trumpeter' in x:
            return False
        if 'trump' in x:
            return True
        return False
    except:
        return False
    
#Read in data and get words. The method for reading in data is one I wrote for Twitter data specifically.
d = tu.loadData('/Users/epierson/Dropbox/scripts/wordsThatDontDescribeHillary/all_tweets', processText = True)
d['words'] = d['tweet_text'].map(getWords)

#Group common adjectives by category so we can get a sense of the most common criticisms. 
#We developed the adjective groups by looking at the most commonly used words,
#grouping them by category and adding words with similar meaning, 
#and then confirming in a random sample of tweets that the ones selected by this filter were appropriate. 

sets_to_examine = {'feminine':['lady like', 'feminine', 'womanly', 'ladylike', 'lady', 'female'], 
                   'nice':['pleasant', 'gracious', 'loved', 'likeable', 'likable', 'warm', 'generous', 'compassionate', 'loving', 'caring', 'kind', 'benevolent', 'nice', 'human', 'relateable', 'relatable'],
                   'physical_pretty':['cutie', 'cute', 'fashionable', 'pretty', 'skinny', 'average looking', 'averagelooking', 'petite','photogenic', 'attractive', 'well-dressed', 'welldressed', 'well dressed', 'beautiful', 'gorgeous','goodlooking', 'good looking', 'cankle less', 'cankleless'],
                   'physical_sexual':['would tap', 'moist', 'gilf', 'sexy', 'hot', 'milf', 'gives good head', 'swallower', 'freaky'],
                   'strong':['strong', 'strength', 'fighter', 'stamina', 'tenacious', 'die hard'],
                   'leader':['potus', 'commander', 'lead', 'president', 'presidential'], 
                   'experienced':['qualified', 'accomplished', 'experienced', 'experience', 'expert', 'capable', 'competent'],
                   'trustworthy':['incorruptible', 'trustworthy', 'lawabiding', 'virtuous', 'forthright', 'moral', 'loyal', 'honor', 'principled', 'consistent', 'transparent','truthful', 'trust', 'authentic', 'integrity', 'innocent', 'honest', 'sincere', 'integrity','accountable', 'decent', 'law abiding', 'law-abiding', 'reliable', 'genuine', 'honest', 'lawful', 'ethical'],
                   'clever':['clever', 'intelligent', 'smart', 'bright', 'canny', 'astute', 'brilliant', 'shrewd', 'forthright'], 
                   'rich':['poor', 'broke', 'middle class', 'wall st', 'will fight wall st'], 
                   'sane':['sober', 'sane', 'stable', 'teetotaler'], 
                   'humble':['humble', 'selfless', 'humility', 'modest'],
                   'american':['patriot', 'patriotic', 'american', 'proamerica', 'pro america'],
                   'liberal':['liberal', 'progressive', 'democrat', 'democratic'], 
                   'feminist':['prowomen', 'feminist', 'pro women', 'intersectional', 'champion of women', 'woman', 'women'],
                   'bill':['bill']
                   }


for wordset in sets_to_examine:
    if wordset != 'bill':#Pretty much any tweet containing Bill refers to Bill Clinton. 
        d[wordset] = d['words'].map(lambda x:1.*any([w in x for w in sets_to_examine[wordset]]))
    else:
        d[wordset] = d['tweet_1grams'].map(lambda x:1.*('bill' in x))
    print 'Number of tweets passing %s filter: %i' % (wordset, d[wordset].sum())#print out number passing each wordset filter. 

d['bernie'] = d['description_1grams'].map(bernie)
d['trump'] = d['description_1grams'].map(trump)

print 'Number of tweets from clear Sanders supporters: %i; number of tweets from clear Trump supporters: %i' % (d['bernie'].sum(), d['trump'].sum())

n_females = len(set(d['screen_name'].loc[d['tweeter_gender'] == 1])) * 1.
n_males = len(set(d['screen_name'].loc[d['tweeter_gender'] == 0])) * 1.
print 'Was able to estimate gender from name for a fraction %2.3f of tweeters' % (1.*(n_females + n_males) / len(set(d['screen_name'])))
print n_females / (n_females + n_males), ' of tweeters with identifiable names had female names'
for col in ['bernie', 'trump']:
    n_females =  len(set(d['screen_name'].loc[(d['tweeter_gender'] == 1) & (d[col] == 1)]))
    n_males =  len(set(d['screen_name'].loc[(d['tweeter_gender'] == 0) & (d[col] == 1)]))
    print 'Identified %i tweeters with female names and %i with male names who passed filter %s' % (n_females, n_males, col)
```

```python
"""
look for differences in content of tweets by gender. 
Statistical significance testing here is pretty fraught because each tweet is clearly not an independent observation. 
1. One tweeter may tweet hundreds of times. 
2. One tweet is retweeted from person to person. 
I discussed these problems with an economics friend who argued that you should dispense with significance testing entirely 
and just report effect sizes since in some sense what you're taking is not a random sample -- it's everyone using the hashtag -- 
and if even you think of it as randomly sampled it's not clear what the sampling process is. 
I take this point but report p-values here nonetheless with an awareness of their caveats. Please take them as heuristics. 
p-values are just the raw unpaired t-test p-values comparing male_list to female_list, where male_list is a list of 
tweets from male tweeters with a 1 if the tweet passes a filter (eg, physical_sexual) and a 0 if it does not. Similarly for female tweeters.
I also report the robust p-values, which cluster standard errors by tweeter.
Note that the effect sizes / significance values for gender, as noted in the piece, are smaller than those for Trump / Sanders affiliation. 
"""

from scipy.stats import ttest_ind
print 'Columns are filter, fraction of tweets passing filter, fraction of women using, fraction of men using, ratio, non-robust p, p'
for wordset in sorted(sets_to_examine, 
                      key = lambda wordset:d[wordset].loc[d['tweeter_gender'] == 1].mean() / d[wordset].loc[d['tweeter_gender'] == 0].mean())[::-1]:
    woman_vals = d[wordset].loc[d['tweeter_gender'] == 1]
    man_vals = d[wordset].loc[d['tweeter_gender'] == 0]
    
    t, p = ttest_ind(woman_vals, man_vals)
    star = '*' if (p < .05) else ''
    idxs = d['tweeter_gender'].map(lambda x:x in [0, 1])
    model = sm.OLS.from_formula('%s ~ tweeter_gender' % wordset, data = d.loc[idxs]).fit()
    robust = model.get_robustcov_results('cluster', groups=d['screen_name'].loc[idxs]) 
    robust_p = robust.pvalues[-1]

    print '%-30s %-5i %2.1f%% women use, %2.1f%% men use, ratio %2.3f, p = %2.3e, robust p = %2.3e %s' % (wordset, d[wordset].sum(), 100 * woman_vals.mean(), 100 * man_vals.mean(), woman_vals.mean() / man_vals.mean(), p, robust_p, star)
```

```python
import statsmodels.api as sm
"""
Instead of comparing tweet content between genders, compare content between Trump/Sanders supporters.
If doSimultaneousRegression is True, run multiple regression controlling for both gender + affiliation simultaneously.
As noted in piece, effect sizes for political party are much larger (and more significant, although this has caveats as discussed above.)
"""
doSimultaneousRegression = True
for wordset in sorted(sets_to_examine, 
                      key = lambda wordset:d[wordset].loc[d['bernie'] == 1].mean() / d[wordset].loc[d['trump'] == 1].mean())[::-1]:
    bernie_vals = d[wordset].loc[(d['bernie'] == 1) ]
    trump_vals = d[wordset].loc[(d['trump'] == 1) ]

    t, p = ttest_ind(bernie_vals, trump_vals)
    star = '*' if (p < .05) else ''
    idxs = d['trump'] | d['bernie']
    model = sm.OLS.from_formula('%s ~ trump' % wordset, data = d.loc[idxs]).fit()
    robust = model.get_robustcov_results('cluster', groups=d['screen_name'].loc[idxs]) 
    robust_p = robust.pvalues[-1]
    print '%-30s %5i %-2.2f%% Sanders use, %2.2f%% Trump use ratio %2.3fx p = %2.3e, robust_p %2.3e %s' % (wordset, d[wordset].sum(), 
                                                                                               100 * bernie_vals.mean(), 100 * trump_vals.mean(), trump_vals.mean() / bernie_vals.mean(), p, robust_p, star)
    
    if doSimultaneousRegression and p < .05:
        df = d.loc[idxs][['screen_name', 'tweeter_gender', wordset, 'trump']].dropna()
        model = sm.OLS.from_formula('%s ~ tweeter_gender + trump' % wordset, data =df).fit()
        robust = model.get_robustcov_results('cluster', groups=df['screen_name'])
        print robust.summary()
    
```
